```python
plt.scatter(date_china,effected_china,label="affected" ,color ='b')
plt.scatter(date_china,death_china,label="death",color ='r')
plt.scatter(date_china,recovered_china,label="recovered",color ='g')
plt.xlabel('dates')
plt.ylabel('no of citizens')
#for y-axis rotation and font size
plt.xticks(fontsize = 7, rotation=45)
plt.title('COVID-19 Statistics in CHINA for first 4 weeks')
#FOR LABELLING COLOUR
plt.legend()
plt.show() 
```
